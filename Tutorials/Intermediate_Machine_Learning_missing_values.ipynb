# **Tutorial**

In this practice work, we will try using following methods to handling the missing values.

1.   Drop columns with missing values
2.   Imputation


```
import pandas as pd
from sklearn.model_selection import train_test_split

# Read the data
X_full = pd.read_csv('train.csv', index_col='Id')
X_test_full = pd.read_csv('test.csv', index_col='Id')
```

```
X_full
```

Below part is perform checking rows (axis=0) for any NaN on "SalePrice" column, then drop the row. y is the copy of "SalePrice" column. Lastly, drop the "SalePrice" column in X_full.

```
# Remove rows with missing target, separate target from predictors
X_full.dropna(axis=0, subset=['SalePrice'], inplace=True)
y = X_full.SalePrice
X_full.drop(['SalePrice'], axis=1, inplace=True)

X_full
```

```
y
```

```
# To keep things simple, we'll use only numerical predictors
X = X_full.select_dtypes(exclude=['object'])
X_test = X_test_full.select_dtypes(exclude=['object'])
```

```
X
```

```
# Break off validation set from training data
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2,
                                                      random_state=0)
```

```
X_train.head()
```

```
# Shape of training data (num_rows, num_columns)
print(X_train.shape)

# Number of missing values in each column of training data
missing_val_count_by_column = (X_train.isnull().sum())
print(missing_val_count_by_column[missing_val_count_by_column > 0])
```

```
missing_val_count_by_column
```

```
# check total rows in the training data
num_rows = X_train.shape[0]
num_rows
```

```
num_cols_with_missing = len(missing_val_count_by_column[missing_val_count_by_column > 0])
tot_missing = sum(missing_val_count_by_column[missing_val_count_by_column > 0])
print(num_cols_with_missing)
print(tot_missing)
```

```
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

# Function for comparing different approaches
def score_dataset(X_train, X_valid, y_train, y_valid):
    model = RandomForestRegressor(n_estimators=100, random_state=0)
    model.fit(X_train, y_train)
    preds = model.predict(X_valid)
    return mean_absolute_error(y_valid, preds)
```

```
# To check the column name which has missing value
column_missing_name = [col for col in X_train.columns if X_train[col].isnull().any()]
column_missing_name
```

# **Drop columns with missing values**

```
reduced_X_train = X_train.drop(column_missing_name,axis=1)
reduced_X_valid = X_valid.drop(column_missing_name,axis=1)

# check the data after drop columns with missing value
reduced_X_train
```

```
print("MAE (Drop columns with missing values):")
print(score_dataset(reduced_X_train, reduced_X_valid, y_train, y_valid))
```

# **Imputation**

Based on documentation from [sklearn](https://scikit-learn.org/stable/modules/generated/sklearn.impute.SimpleImputer.html). Imputation transformer for completing missing values. It has few imputation strategy.

The imputation strategy.

* If â€œ**mean**â€, then replace missing values using the mean along each column. Can only be used with numeric data.

* If â€œ**median**â€, then replace missing values using the median along each column. Can only be used with numeric data.

* If â€œ**most_frequent**â€, then replace missing using the most frequent value along each column. Can be used with strings or numeric data.

* If â€œ**constant**â€, then replace missing values with fill_value. Can be used with strings or numeric data.

```
from sklearn.impute import SimpleImputer

my_imputer = SimpleImputer()
imputed_X_train = pd.DataFrame(my_imputer.fit_transform(X_train))
imputed_X_valid = pd.DataFrame(my_imputer.transform(X_valid))
```

```
imputed_X_train
```

```
# after imputation, the column names were removed, therefore need to assign the column name
imputed_X_train.columns = X_train.columns
imputed_X_valid.columns = X_valid.columns
```

```
imputed_X_train
```

```
print("MAE (Imputation):")
print(score_dataset(imputed_X_train, imputed_X_valid, y_train, y_valid))
```

Surprisingly, the method of dropping column is perform slightly better than imputation method. This may because of:

1.  There is noise in the dataset
2.  The data is not suitable to using imputation since not every missing data is consider approach to mean value.

```
final_imputer = SimpleImputer(strategy='median')
# Preprocessed training and validation features
final_X_train = pd.DataFrame(final_imputer.fit_transform(X_train))
final_X_valid = pd.DataFrame(final_imputer.transform(X_valid))
```

```
type(final_X_train)
```

```
final_X_train.columns = X_train.columns
final_X_valid.columns = X_valid.columns
```

```
model = RandomForestRegressor(n_estimators=100, criterion='mse', random_state=0)
model.fit(final_X_train, y_train)

pred_valid = model.predict(final_X_valid)
print("MAE: ")
print(mean_absolute_error(y_valid, pred_valid))
```
